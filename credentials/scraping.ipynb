League Info: Premier League

```python
url = 'https://www.transfermarkt.com/vereins-statistik/wertvollstemannschaften/marktwertetop/plus/0/galerie/0?land_id=189&kontinent_id=6&yt0=Show'
browser = Browser('chrome', headless = True)
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
all_leagues = []
```

```python
table = soup.find('table',{'class':'items'})
table_details = table.find('tbody').find_all('tr')
```

```python
team_list = []
value_list = []
country = []
league_name = []

for table in table_details:
    row_a = table.find_all('a')
    if row_a[2]['title'] == 'Premier League':
        league_name.append(row_a[2]['title'])
        team_list.append(row_a[1]['title'])
        row_b = table.find('b')
        value_list.append(row_b.text)
        row_img = table.find_all('img')
        country.append(row_img[1]['title'])
```

```python
url_second = 'https://www.transfermarkt.us/premier-league/eigenetabelle/wettbewerb/GB1'
browser = Browser('chrome', headless = True)
browser.visit(url_second)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
table = soup.find('table',{'class':'items'})
table_details = table.find('tbody').find_all('tr')
```

```python
club_list = []
games_played= []
win = []
draw = []
lose = []
goals = []
points = []
points_per_match = []
for table in table_details:
    row_a = table.find('a')
    row_games = table.find_all('td',{'class':'zentriert'})
    club_list.append(row_a['title'])
    games_played.append(row_games[2].text)
    win.append(row_games[3].text)
    draw.append(row_games[4].text)
    lose.append(row_games[5].text)
    goals.append(row_games[6].text)
    points.append(row_games[8].text)
    points_per_match.append(row_games[9].text)
```

```python
premier_league_one = pd.DataFrame(list(zip(team_list, country, league_name, value_list,)), columns = [
                            'Club', 'Country', 'League', 'Market Value'])

premier_league_two = pd.DataFrame(list(zip(club_list, games_played, win, draw, lose, goals, points, points_per_match)), 
                                  columns = ['Club', 'PL', 'W', 'D', 'L', 'G', 'Pts', 'Pts_per_match'])

premier_league = pd.merge(premier_league_one, premier_league_two, on='Club')
```

```python
premier_league
```

```python
all_leagues.append(premier_league)
```
